```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
# C:\Users\cuong\Downloads\avg_rain_state.csv
# http://localhost:8888/edit/OneDrive/DataScience/VBA_Excel/avg_rain_state%20(1).csv
# C:\Users\cuong\OneDrive\DataScience\VBA_Excel\avg_rain_state (1).csv
rain_df = pd.read_csv("C:/Users/cuong/OneDrive/DataScience/VBA_Excel/avg_rain_state (1).csv")
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(rain_df))
tick_locations = [value + 0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(9,5))
plt.bar(x_axis, rain_df["Inches"], color ="g", alpha=0.8, align="edge")
plt.xticks(tick_locations, rain_df["State"], rotation ="vertical")
```

```python
# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(rain_df["Inches"])+10)
```

```python
# Set a Title and labels
plt.title("Average Rain State")
plt.xlabel("State")
plt.ylabel("Average Amount of Rainfall in Inches")
```

```python
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("avg_state_rain.png")
plt.show()
```
